## 👉**Exercise 1 : Implement the Hamming** 

As a reminder, the **Hamming Distance** measures the **similarity between 2 strings of the same length.** The Hamming Distance between 2 strings of the same length is the number of positions at which
the corresponding characters are different. This distance can be also expressed as the proportion of disagreeing characters between 2 strings (position of characters is important).

- Write a function called *dist_hamming(string1, String2)* which returns the hamming distance as tuple consisting of (the number of positions at which the characters are different, the proportion of disagreeing characters between both strings) if strings have the same length. If not,  Your function must return this message: "Cannot compute the hamming distance: length of strings are different"

- Test your function on the couples of strings below: 

        - ALEXANDRE and ALEKSANDR

		- CLOSE and CLOUE

		- LIONGMOT and LONGMOIT 

		- LONGIMOT and LONGMOIT 

		- WIKIMEDIA and WIKIMANIA	

		- ALEXANDRE and ALEKSANDER
		
		- WIKIMEDIA and REMEDY
	

```python
# YOUR CODE HERE
def dist_hamming(string1, string2):
    if len(string1) != len(string2):
        return "Cannot compute the Hamming distance: the lengths of strings are different"
    
    differences = 0
    
    for c1, c2 in zip(string1, string2):
        if c1 != c2:
            differences += 1
    
    proportion = differences / len(string1)
    return (differences, proportion)

tests = [("ALEXANDRE", "ALEKSANDR"), ("CLOSE", "CLOUE"), ("LIONGMOT", "LONGMOIT"), ("LONGIMOT", "LONGMOIT"), ("WIKIMEDIA", "WIKIMANIA"), ("ALEXANDRE", "ALEKSANDER"), ("WIKIMEDIA", "REMEDY")]

for s1, s2 in tests:
    result = dist_hamming(s1, s2)
    print(f'The Hamming distance between "{s1}" and "{s2}" is: {result}')
    print("*" * 100)
```

## 👉**Exercice 3 : Jaccard Similarity**
As a reminder, Jaccard Similarity is defined as an intersection of two documents divided by the union of that two documents that refer to the number of common words over a total number of words. 
The mathematical representation of the Jaccard Similarity is:
$$Jaccard(doc_1,doc_2)= \frac{doc_1 \cap doc_2}{doc_1 \cup doc_2}$$

- Write a python program computing the Jaccard similarity 

- Test your code with the sentences :

    - Document_1="Data is the new oil of the digital economy"

    - Document_2="Data is a new oil"

```python
# YOUR CODE HERE
def Jaccard_Similarity(document1, document2): 
    
    # List the unique words in a document
    words_doc1 = set(document1.lower().split()) 
    words_doc2 = set(document2.lower().split())
    
    # Find the intersection of words list of document1 & document2
    intersection = words_doc1 & words_doc2

    # Find the union of words list of document1 & document2
    union = words_doc1 | words_doc2
        
    # Calculate Jaccard similarity score
    return len(intersection) / len(union)
```

```python
# YOUR CODE HERE
#Test your function on both document
Document_1= "Data is the new oil of the digital economy"
Document_2= "Data is a new oil"
score = Jaccard_Similarity(Document_1, Document_2)
print(score)
```

 In this exercise, you **are given a misspelled word**, and you need **to find the most likely correct word from a list of words** using the Levenshtein distance. As a reminder the Levenshtein distance is used to measure the number of edits (insertions, deletions, or substitutions) required to convert one word (source) into another (target). This distance is **useful for detecting typos and suggesting corrections.**

1. Implement a function called **spell_correction** satisfying this:
     - take as argument a **misspelled_word** (representing the misspelled word) and the **list of correct words**
    - Create a **list of tuples** where each tuple consists of **each correct word and its Levenshtein distance to the misspelled word**
    - **Sort this list by ascending distance**
    - Return **the word with the smallest distance**  

2. Test your function with the misspelled word: "compter".



```python
import Levenshtein

# List of correct words
List_of_correct_words = ["cat", "dog", "house", "car", "computer", "machine", "artificial", "intelligence"]

# Function to find the closest suggestion
def spell_correction(misspelled_word, List_of_correct_words):
    # list of tuples: (word, distance)
    distances = []
    
    for word in List_of_correct_words:
        d = Levenshtein.distance(misspelled_word, word)
        distances.append((word, d))
    
    # Sort by ascending distance
    distances.sort(key=lambda x: x[1])
    
    # Return the word with the smallest distance
    return distances[0][0]

# Misspelled word
incorrect_word = "compter"   # 'u' missing in "computer"

# Find the correction: test your function
correction = spell_correction(incorrect_word, List_of_correct_words)

print("Misspelled word:", incorrect_word)
print("Suggested correction:", correction)
```

👉 Now, let us define a function called correct_sentence satisfying this:   Given a sentence, the function check each word for spelling mistakes and suggest corrections. This function uses the function spell checker defined above.

```python
import nltk
from nltk.corpus import words
from nltk.metrics.distance import edit_distance

# Download NLTK words dictionary
nltk.download("words")

# Dictionary
dictionary = set(word.lower() for word in words.words())

# Spell checker (top 5 suggestions)
def spell_checker(misspelled_word, dictionary, top_n=5):
    distances = [(word, edit_distance(misspelled_word.lower(), word.lower())) for word in dictionary]
    distances.sort(key=lambda x: x[1])
    return [word for word, _ in distances[:top_n]]

# Correct sentence function
def correct_sentence(sentence, dictionary, top_n=5):
    words_in_sentence = sentence.split()
    corrected_sentence = []

    for word in words_in_sentence:
        clean_word = word.lower()

        if clean_word in dictionary:
            corrected_sentence.append(word)
        else:
            suggestions = spell_checker(clean_word, dictionary, top_n)
            corrected_sentence.append(f"[{word} -> {', '.join(suggestions)}]")

    return " ".join(corrected_sentence)

# TEST
input_sentence = "This is a smple sentnce with soome mistake"
corrected = correct_sentence(input_sentence, dictionary)

print("Corrected Sentence:")
print(corrected)
```

```python
#exemple 1
from numpy import dot
from numpy.linalg import norm
def cosine_similarity1(a,b):
    return np.dot(a, b) / (norm(a) * norm(b))

# Vecteur donnés

X = np.array([2,3,4])
Y = np.array([1,3,4])

#Result
result = cosine_similarity1(X, Y)

print(result)
```

## **Another way**

```python
#exemple 2
def cosine_similarity2(a, b):
    nominator = np.dot(a, b)
    
    a_norm = np.sqrt(np.sum(a**2))
    b_norm = np.sqrt(np.sum(b**2))
    
    denominator = a_norm * b_norm
    
    cosine_similarity = nominator / denominator
    
    return cosine_similarity

print(cosine_similarity2(X,Y))
```

```python

# note that this function actually calculates cosine similarity 
# and then use "1-similarity" to convert similarity to distance
# to get the actual cosine similarity, you need to do 1-distance

from scipy import spatial
import numpy as np

def cosine_similarity3(a, b):
    return 1 - spatial.distance.cosine(a, b)

X = np.array([2, 3, 4])
Y = np.array([1, 3, 4])

result = cosine_similarity3(X, Y)

print("Cosine similarity (scipy):", result)
```

0.30 – 0.49 (30% – 49%)	🟢 Low similarity → Likely coincidence, related topics but different writing.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download stopwords from NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Define the documents
document_a = """
In recent years, artificial intelligence (AI) has become a pivotal part of many industries. 
From healthcare to finance, AI has revolutionized the way we approach problems and offer solutions. 
Machine learning, a subset of AI, has enabled computers to learn from data, improving the efficiency of systems. 
With applications in autonomous vehicles, personalized medicine, and data analytics, AI continues to grow at an exponential rate. 
The future of AI holds significant promise, and we are just beginning to scratch the surface of its potential.
"""

document_b = """
The rapid growth of artificial intelligence (AI) in recent years has transformed multiple sectors, including healthcare and finance. 
AI, and its subset machine learning, have enabled systems to analyze data and improve performance. 
Machine learning allows computers to adapt and optimize their actions based on data analysis. 
Industries like autonomous vehicles, personalized healthcare, and data analysis are already benefiting from the rapid advancements in AI. 
As AI continues to evolve, it holds enormous potential for the future, and we are still in the early stages of exploring its capabilities.
"""

# Step 1: Preprocess the documents (remove stopwords and tokenize)
stop_words = set(stopwords.words("english"))

def preprocess(text):
    tokens = text.lower().split()
    filtered_tokens = [
        word for word in tokens
        if word.isalnum() and word not in stop_words
    ]
    return " ".join(filtered_tokens)

doc_a_clean = preprocess(document_a)
doc_b_clean = preprocess(document_b)

# Step 2: Vectorize the documents using TF-IDF
vectorizer = TfidfVectorizer()

# Combine the documents into a single list for vectorization
documents = [doc_a_clean, doc_b_clean]
tfidf_matrix = vectorizer.fit_transform(documents)

# Step 3: Calculate the cosine similarity between the two documents
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

# Step 4: Output the result
print("Document A nettoyé :")
print(doc_a_clean)
print("\nDocument B nettoyé :")
print(doc_b_clean)

print("\nCosine Similarity entre Document A et Document B :")
print(round(similarity, 4))

# 6. Interprétation
if similarity >= 0.90:
    interpretation = "Très forte similarité : suspicion élevée de plagiat."
elif similarity >= 0.75:
    interpretation = "Similarité importante : possible reformulation / paraphrase."
elif similarity >= 0.50:
    interpretation = "Similarité modérée : mêmes thèmes ou vocabulaire proche."
elif similarity >= 0.30:
    interpretation = "Faible similarité : quelques points communs seulement."
else:
    interpretation = "Très faible similarité : documents plutôt différents."

print("\nInterprétation :")
print(interpretation)
```

```python
#Cosine similarity is used to measure the closeness between two documents after preprocessing and TF-IDF vectorization. The closer the score is to 1, the more similar the documents are. In this case, the score obtained helps determine whether there is strong suspicion of plagiarism or merely thematic similarity.
```

Discussion of Cosine Similarity:

Cosine similarity is widely used in antiplagiarism detection because it is effective in measuring the similarity of text documents while ignoring differences in text length. It works well when there are minor modifications (e.g., synonym substitution or slight rephrasing). However, it may not capture deep semantic changes or more complex forms of plagiarism where the structure and flow of the content are significantly altered.

- For information, The Mahalanobis distance accounts for the covariance structure of the data and adjusts for the scale and orientation of the data. It is defined as:


$$
D_M(\mathbf{x}, \mathbf{y}) = \sqrt{(\mathbf{x} - \mathbf{y})^T \mathbf{S}^{-1} (\mathbf{x} - \mathbf{y})}
$$

where:  
- $\mathbf{x}$  and $\mathbf{y}$ are vectors representing two points in the dataset.  
- $\mathbf{S}$ is the covariance matrix of the dataset.  
- $\mathbf{S}^{-1}$ is the inverse of the covariance matrix.  
- $(\mathbf{x} - \mathbf{y})^T$ is the transpose of the difference vector.  



By using the Mahalanobis distance, the effect of outliers can be reduced because it considers the spread and orientation of the data.

The **Mahalanobis distance** between a point $\mathbf{x}$ and a distribution with mean $\boldsymbol{\mu}$ and covariance matrix $\mathbf{\Sigma}$ is given by the formula (the one used below):

$$
D_M(\mathbf{x}) = \sqrt{(\mathbf{x} - \boldsymbol{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu})}
$$

where:  
- $\mathbf{x}$ is the data point (a vector).  
- $\boldsymbol{\mu}$ is the mean vector of the distribution.  
- $\mathbf{\Sigma}$ is the covariance matrix of the distribution.  
- $\mathbf{\Sigma}^{-1}$ is the inverse of the covariance matrix.  
- $(\mathbf{x} - \boldsymbol{\mu})^T$ represents the transpose of the difference vector.  

This distance measures how far $\mathbf{x}$ is from the mean, taking into account correlations between variables. It is widely used in multivariate anomaly detection and classification.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import pandas as pd

# Import the dataset 


# Compute the covariance matrix of the dataset
cov_matrix = np.cov(data.T)

# Compute the inverse of the covariance matrix
cov_inv = np.linalg.inv(cov_matrix)
# Compute the Mahalanobis distances of each point from the mean
mahalanobis_distances = np.array([distance.mahalanobis(point, np.mean(data, axis=0), cov_inv) for point in data])

# Compute the centroid of the dataset using Mahalanobis distance
mahalanobis_weights = 1 / (mahalanobis_distances + 1e-8)  # Adding a small value to avoid division by zero
mahalanobis_centroid = np.average(data, axis=0, weights=mahalanobis_weights)

euclidean_centroid = np.mean(data, axis=0)
print (mahalanobis_centroid)
print (euclidean_centroid)
# Plot the dataset and the Mahalanobis centroid
plt.scatter(data[:, 0], data[:, 1], label='Data Points')
plt.scatter(euclidean_centroid[0], euclidean_centroid[1], color='red', marker='x', label='Euclidean Centroid')
plt.scatter(mahalanobis_centroid[0], mahalanobis_centroid[1], color='green', marker='o', label='Mahalanobis Centroid')
plt.title("Dataset with Centroids")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
